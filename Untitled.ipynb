```python
mask = (data['time'] > 30)&(data['time'] < data['time'][-1]-80)
```

```python
plt.figure(figsize=(12, 3))
plt.plot( data['time'][mask], data['Gain'][mask], 'r-' )
plt.xlabel('time [s]');
plt.xlim( min(data['time'][mask]), max(data['time'][mask]) )
plt.ylabel('Gain');
```

```python
plt.figure(figsize=(12, 6))
plt.subplot( 4, 1, 1 )
plt.plot( data['time'][mask], data['gFx'][mask], '-' )
plt.ylabel('gFx');
plt.xlim( min(data['time'][mask]), max(data['time'][mask]) )
plt.tick_params(labelbottom='off')

plt.subplot( 4, 1, 2 )
plt.plot( data['time'][mask], data['gFy'][mask], '-' )
plt.ylabel('gFy');
plt.xlim( min(data['time'][mask]), max(data['time'][mask]) )
plt.tick_params(labelbottom='off')

plt.subplot( 4, 1, 3 )
plt.plot( data['time'][mask], data['gFz'][mask], '-' )
plt.xlabel('time (s)');
plt.xlim( min(data['time'][mask]), max(data['time'][mask]) )
plt.ylabel('gFz');
plt.tick_params(labelbottom='off')


gFx = data['gFx'][mask]
gFy = data['gFy'][mask]
gFz = data['gFz'][mask]
gFnorm = np.sqrt( gFx**2 + gFy**2 + gFz**2  )

plt.subplot( 4, 1, 4 )
plt.axhline( y=1, color='k', linewidth=1 )
plt.plot( data['time'][mask], gFnorm, 'r-' )
plt.xlabel('time (s)');
plt.xlim( min(data['time'][mask]), max(data['time'][mask]) )
plt.ylabel('norm');
```

```python
plt.figure(figsize=(12, 5))
plt.subplot( 3, 1, 1 )
plt.plot( data['time'][mask], data['Azimuth'][mask], 'm-' )
plt.ylabel('Azimuth');
plt.xlim( min(data['time'][mask]), max(data['time'][mask]) )
plt.tick_params(labelbottom='off')

plt.subplot( 3, 1, 2 )
plt.plot( data['time'][mask], data['Pitch'][mask], 'r-' )
plt.ylabel('Pitch');
plt.xlim( min(data['time'][mask]), max(data['time'][mask]) )
plt.tick_params(labelbottom='off')

plt.subplot( 3, 1, 3 )
plt.plot( data['time'][mask], data['Roll'][mask], '-' )
plt.xlabel('time (s)');
plt.xlim( min(data['time'][mask]), max(data['time'][mask]) )
plt.ylabel('Roll');
```
